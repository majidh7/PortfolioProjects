```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.opencodez.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text , 'html.parser')
latest = soup.find('div' , attrs = {'class' : 'latest-section'})
```

```python
last_page = int(soup.findAll('a' , attrs = {'class' : 'inactive'})[-1]['href'].split('/')[-1])

titles = []
descriptions = []
authors = []
years = []
month_days = []
links = []

for p in range(1, last_page + 1):
    url = 'https://www.opencodez.com/page/{}'.format(p)
    response = requests.get(url)
    soup = BeautifulSoup(response.content , 'html.parser')
    latest = soup.find('div' , attrs = {'class' : 'latest-section'})
    
    title = latest.findAll('h2' , attrs = {'class' : 'title'})
    description = latest.findAll('div' , attrs = {'class' : 'post-content image-caption-format-1'})
    author = latest.findAll('span' , attrs = {'class': 'theauthor'})
    year = latest.findAll('span' , attrs = {'class' : 'thetime'})
    month_day = latest.findAll('span' , attrs = {'class' : 'thetime'})
    link = link = latest.findAll('h2' , attrs = {'class' : 'title'})

    each_list_len = [len(title) , len(description) , len(author) , len(year) , len(month_day) , len(link)]
    
    if len(set(each_list_len)) == 1:
        for i in range(0, each_list_len[0]):
            titles.append(title[i].find('a').get_text())
            descriptions.append(description[i].find('p').get_text().strip())
            authors.append(author[i].find('a').get_text())
            years.append(year[i].get_text().strip().split(',')[1])
            month_days.append(month_day[i].get_text().strip().split(',')[0])
            links.append(link[i].find('a').get('href'))
    else:
        print('null')
        
    print('Scraping page {} is finished.'.format(p))
```

```python
data = pd.DataFrame(list(zip(titles , descriptions , authors , years , month_days , links)) ,
                   columns = ['Title' , 'description' , 'Author' , 'Years' , 'Month_days' , 'Link'])
```
